```python
import scanpy as sc
import matplotlib.pyplot as plt
import numpy as np
```

```python
lat_infected = sc.read_h5ad('V105_infected_Neutrophils_with_umap.h5ad')
lat_controls = sc.read_h5ad('V105_control_Neutrophils_with_umap.h5ad')
```

```python
#palette_1 = ['#FF7878', '#8B0000', '#3498DB', '#1F618D']
#palette_1 = ['#00FFFF', '#8B0000', '#3498DB', '#1F618D']
palette_1 = ['#e27d26ff', '#C63735', '#62205fff', '#4e3910ff']
#00FFFF
#palette_2 = ['#FFA5B1', '#AED6F1']
palette_2 = ['#1a71b8ff','#4fb6caff']
```

```python
fig, axes = plt.subplots(1, 2, figsize=(10, 4))
# Plot UMAP for AnnData 2
sc.pl.umap(lat_controls, ax=axes[0],show = False,color = 'compare_umap',palette = palette_2,size = 3*np.ones(len(lat_controls)),alpha = 0.7)
axes[0].set_title('Controls')
axes[0].legend(loc='center left', bbox_to_anchor=(-0.4, 0.5))
#axes[0].invert_xaxis()
# Plot UMAP for AnnData 1
sc.pl.umap(lat_infected,color = 'compare_umap',ax=axes[1],show = False,palette = palette_1,size = 3*np.ones(len(lat_infected)),alpha = 0.7)
axes[1].set_title('Infection Cases')
axes[1].legend(loc='center left', bbox_to_anchor=(1, 0.5))

# Adjust spacing between subplots
plt.tight_layout()
plt.savefig('V105_5_show_shift_v1.pdf')
```
